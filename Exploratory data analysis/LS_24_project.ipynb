```python
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv('creditcard_for_fraud_detection.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
```

```python
for cols in range(1, 29):
    plt.scatter(x = df.iloc[:, cols], y = df.iloc[:, 29])
    plt.xlabel(f'Column {cols}')
    plt.ylabel('Column 29')
    plt.title(f'Scatter Plot of Column {cols} vs Column 29')
    plt.show()
```

```python
clf = IsolationForest(contamination = 0.00172748)
```

```python
clf.fit(df)
```

```python
predict = clf.predict(df)
```

```python
predict
```

```python
import numpy as np
abd_index = np.where(predict<0)
```
