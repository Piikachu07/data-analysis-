```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv("train.csv")
```

```python
flights = pd.read_csv('AirPassengers.csv')
```

# 1.scatterlpot (Numerical-Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', data = tips, hue=tips['sex'], style=tips['smoker'], size = tips['size'])
```

# 2.Bar Plot (Numerical-Categorical) 

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic, hue = titanic['Sex'])
```

# 3.Boxplot(Numerical-Categorical)

```python
sns.boxplot(x='Sex',y='Age', data=titanic, hue=titanic['Survived'],saturation=1)
```

# 4.displot (Numerical-Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False,color='Red',rug=True)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False,)
```

# 5.Heatmap(Categorical-categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```
